```python
# Set the folder path for audio files
folder_path = "C:\\Users\\Dev\\Documents\\SpotifAI Audio Files Dataset"

# Create an empty list to store DataFrames
all_chroma_dfs = []

# Loop through audio files in the specified folder
for audio_file in os.listdir(folder_path):
    if audio_file.endswith('.mp3'):
        audio_path = os.path.join(folder_path, audio_file)

        try:
            # Load the audio file
            x, sr = librosa.load(audio_path)

            # Calculate chroma features
            chroma = librosa.feature.chroma_stft(y=x, sr=sr)

            # Calculate mean and standard deviation of chroma
            chroma_mean = np.mean(chroma, axis=1)
            chroma_std = np.std(chroma, axis=1)

            # Generate the chroma DataFrame
            chroma_df = pd.DataFrame()
            for i in range(0, 12):
                chroma_df['chroma_mean_' + str(i)] = chroma_mean[i]
            for i in range(0, 12):
                chroma_df['chroma_std_' + str(i)] = chroma_std[i]
            chroma_df.loc[0] = np.concatenate((chroma_mean, chroma_std), axis=0)

            # Add a column for the audio file name
            chroma_df['Audio File'] = audio_file

            # Append the DataFrame to the list
            all_chroma_dfs.append(chroma_df)

        except Exception as e:
            print(f"Error loading {audio_file}: {e}")

# Concatenate all DataFrames into a single DataFrame
final_chroma_df = pd.concat(all_chroma_dfs, ignore_index=True)

# Display the final DataFrame
print(final_chroma_df)
```

```python
# Define the folder where you want to save the final CSV file on your computer
output_folder = "C:\\Users\\Dev\\Documents\\SpotifAI_Chroma_Data"
os.makedirs(output_folder, exist_ok=True)


# Define the output path and save the final DataFrame to a CSV file
output_file_path = os.path.join(output_folder, "combined_chroma_data.csv")
final_chroma_df.to_csv(output_file_path, index=False)

# Display the final DataFrame and a confirmation message
#print(final_chroma_df)
#print(f"All chroma data saved to {output_file_path}")
```

```python

import os
import numpy as np
import pandas as pd
import librosa

# Set the folder path to your local directory with audio files
folder_path = "C:\\Users\\Dev\\Documents\\SpotifAI Audio Files Dataset"

# Define the folder where you want to save the final CSV file on your computer
output_folder = "C:\\Users\\Dev\\Documents\\SpotifAI_MFCC_Data"
os.makedirs(output_folder, exist_ok=True)

# Initialize an empty DataFrame to store all MFCC data
all_mfccs_df = pd.DataFrame()

# Loop through audio files in the specified folder
for audio_file in os.listdir(folder_path):
    if audio_file.endswith('.mp3'):
        audio_path = os.path.join(folder_path, audio_file)
        
        try:
            # Load the audio file using librosa
            x, sr = librosa.load(audio_path)
            print(f"Loaded {audio_file}")

            # Calculate MFCCs
            mfccs = librosa.feature.mfcc(y=x, sr=sr, n_mfcc=13)
            mfccs_mean = np.mean(mfccs, axis=1)
            mfccs_std = np.std(mfccs, axis=1)

            # Create a dictionary for the current file's MFCC data
            mfcc_data = {'File': audio_file}
            for i in range(13):
                mfcc_data[f'mfccs_mean_{i}'] = mfccs_mean[i]
                mfcc_data[f'mfccs_std_{i}'] = mfccs_std[i]

            # Convert the dictionary to a DataFrame and concatenate it to the main DataFrame
            mfcc_df = pd.DataFrame([mfcc_data])
            all_mfccs_df = pd.concat([all_mfccs_df, mfcc_df], ignore_index=True)

        except Exception as e:
            print(f"Error loading {audio_file}: {e}")

# Save the combined DataFrame to a CSV file in the specified output folder
output_file_path = os.path.join(output_folder, "combined_mfcc_data.csv")
all_mfccs_df.to_csv(output_file_path, index=False)
print(f"All MFCC data saved to {output_file_path}")
```
